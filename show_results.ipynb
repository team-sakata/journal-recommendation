## 基本的なSciSci論文ラベル付のアルゴリズム
1. SciSci専門誌の論文の引用・被引用関係にある論文で、一定数(=scisci_ratio)以上のSciSci論文を引用・被引用している論文をSciSci論文として取得。
2. 1を繰り返すことで、バームクーヘンのようにSciSci関連論文を取得できる

## How to use
- scisci_ratioのtheresholdを調整して、読みたい論文を出力してみてください。
- [10, 15, 20, 25, 30, 35, 40]倍の結果を格納しています。

```python
import pickle
import pandas as pd
import random
```

## 検証用のScopusデータの読み込み

```python
papers_data = pd.concat([
  pd.read_pickle('/disks/qnap3/shared/scopus-24/data/paper/eid.pickle'),
  pd.read_pickle('/disks/qnap3/shared/scopus-24/data/paper/num_cited.pickle'),
  pd.read_pickle('/disks/qnap3/shared/scopus-24/data/paper/doi.pickle'),
  pd.read_pickle('/disks/qnap3/shared/scopus-24/data/paper/title.pickle'),
  pd.read_pickle('/disks/qnap3/shared/scopus-24/data/paper/journal.pickle'), 
  # pd.read_pickle('/disks/qnap3/shared/scopus-22/data/paper_detail/authids.pickle'),
  pd.read_pickle('/disks/qnap3/shared/scopus-24/data/paper/year.pickle'),
], axis = "columns")
print(papers_data.head())
```

## corepapersをscisci専門誌にした場合
閾値: [10,15,20, 25, 30, 35, 40] * scisci_ratio

### 出力結果の読み込み

```python
with open('/disks/qnap3/users/21-tomokiyo/journal-recommendation/result/output_corepaper=only_specialized_journals/result1', 'rb') as f:
    output_dict = pickle.load(f)
```

### 閾値の指定

```python
threshold = 10
result_eid = output_dict[threshold] #resultにはeidが格納
```

### ランダムに結果を10個出力
ジャーナルの指定なし

```python
output_eid = random.sample(result_eid, 20) #シャッフルしてeidを出力

print('Threshold:', threshold)
papers_data.loc[list(output_eid)] 
```

### トップ総合誌で結果を出力

```python
target_journal =  [21100838541, # Nature Human Behaviour
  19700182758, # Nature Communications
  23571, # Science
  21121, #PNAS
  21206 #Nature
]
target_papers_subset = papers_data[papers_data["journal"].isin(target_journal)].fillna(0)
target_papers_eid = target_papers_subset.eid.to_list()
output_eid = [i for i in target_papers_eid if i in result_eid]
```

```python
output_papers_data = target_papers_subset[target_papers_subset['eid'].isin(output_eid)]
```

```python
output_papers_data.head(20)
```

```python
# ランダムに出力
output_eid = random.sample(output_eid, 20) #シャッフルしてeidを出力

print('Threshold:', threshold)
target_papers_subset.loc[list(output_eid)] 
```
