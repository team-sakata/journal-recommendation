```python
from dotenv import load_dotenv
# from util import cluster_topics, visualize_clusters
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
# import re
import os

load_dotenv()

data_path = os.getenv('DATA_PATH')
result_path = os.getenv('PROCESSED_PATH')
figure_path = os.getenv('FIG_PATH')
rawdata_path = os.getenv('RAW_PATH')
```

```python
asjcs_df = pd.read_pickle(data_path+"paper_detail/asjcs.pickle")
asjc_df = asjcs_df.explode()
# cs = asjc_df[asjc_df%1700<100] # five times slower than boolean multiplication
cs = asjc_df[(asjc_df>=1700) * (asjc_df<1800)]
cs.head()
```

```python
citation_df = pd.read_pickle(data_path+"citations_gb.pickle")
journal_df =  pd.read_pickle(data_path+"paper_detail/journal.pickle")
citation_cs = citation_df[citation_df.index.isin(cs.index)]
citation_cs.head()
```

```python
cs_journal_pub = journal_df[journal_df.index.isin(cs.index.values)]
cs_journal_pub.name="journal"
cs_journal_pub = pd.merge(cs_journal_pub,citation_cs, left_index=True, right_index=True)

journal_source_target = pd.merge(cs_journal_pub[["journal","target"]].explode("target")
                                 ,journal_df
                                 ,left_on="target",right_index=True)\
                        .rename(columns={"journal_x":"source_journal", "journal_y":"target_journal"})
journal_source_target.head()
```

```python
journal_citation_weights = journal_source_target.groupby(by="source_journal").value_counts(subset=["target_journal"])
journal_citation_weights = journal_citation_weights[journal_citation_weights>10].reset_index()

# backward compatibility in value_counts
if not "count" in journal_citation_weights.columns:
    journal_citation_weights.columns = ["source_journal", "target_journal", "count"]

journal_citation_weights.shape[0], journal_citation_weights.nunique(), journal_citation_weights.head()
```

```python
weights_matrix = journal_citation_weights.pivot(index='source_journal', columns='target_journal', values='count').fillna(0)

print("median citation from jounal to journal:", weights_matrix.max().median()
      , "max citation from jounal to journal:",weights_matrix.max().max())

# plt.imshow(weights_matrix,vmax=weights_matrix.max().median(),vmin=0)
plt.imshow(weights_matrix,vmax=1,vmin=0) # journal-journal citation is sparse

array = weights_matrix.to_numpy()
print(array.shape) # before omission
array = array[:, ~(array==0).all(axis=0)]
print(array.shape) # after omission... no change
# this means there is at least one journal citing more than 10 times for every journal.
```

```python
# target_journals = weights_matrix.columns
# len(weights_matrix.columns), len(set(weights_matrix.columns) - set(weights_matrix.index)) + len(weights_matrix.index)

weights_matrix_winsorized = weights_matrix.loc[weights_matrix.sum(axis=1)>100,weights_matrix.max(axis=0)>30]
weights_matrix_normalized = weights_matrix_winsorized.div(
    np.linalg.norm(weights_matrix_winsorized.values, axis=1)
    , axis=0)
```

```python
cosine_similarity = np.dot(weights_matrix_normalized.values, weights_matrix_normalized.values.T)
plt.imshow(cosine_similarity,cmap="viridis",vmax=0.1,vmin=0)
```

```python
plt.imshow(weights_matrix_normalized,cmap="magma_r",vmax=0.01,vmin=0) # journal-journal citation is sparse
```

```python
mean = weights_matrix_winsorized.mean(axis="index")
sigma = np.sqrt(weights_matrix_winsorized.var(axis="index"))

norms = weights_matrix_winsorized.sub(mean,axis="columns")
norms = norms.div(sigma, axis="columns")
plt.imshow(norms, cmap="coolwarm", vmax=0.1, vmin=-0.1)
```

```python
weights_matrix_normalized_l1 = weights_matrix_winsorized.div(weights_matrix_winsorized.sum(axis="columns"), axis="index")

mean = weights_matrix_normalized_l1.mean(axis="index")
sigma = np.sqrt(weights_matrix_normalized_l1.var(axis="index"))

norms = weights_matrix_normalized_l1.sub(mean,axis="columns")
norms = norms.div(sigma, axis="columns")
plt.imshow(norms, cmap="coolwarm", vmax=0.1, vmin=-0.1) # some of the columns are all np.nan

norms_valid = norms.loc[:,~norms.isna().any(axis=0)] 
norms.sum(axis="columns")
```

```python
norms_l2 = norms_valid.div(np.linalg.norm(norms_valid.values, axis=1),axis="index")

cosine_similarity = np.dot(norms_l2.values, norms_l2.values.T)
plt.imshow(cosine_similarity,cmap="viridis",vmax=0.1,vmin=-0.01)
# plt.imshow(cosine_similarity,cmap="viridis",vmax=np.median(cosine_similarity.max()),vmin=np.median(cosine_similarity.min()))
```

```python
order = np.argsort(cosine_similarity.sum(axis=0))
print(order)
clustered_adj_matrix = cosine_similarity[order[::-1], :][:, order[::-1]]

plt.imshow(clustered_adj_matrix,cmap="viridis",vmax=0.1,vmin=-0.01)
```

```python
# L2-normalize
import pandas as pd
import numpy as np

# 
data = {
    'A': [100, 2, 3],
    'B': [10000, 101, 99],
    'C': [700, 8, 9]
}
# data = {
#     'A': [1, 2, 3],
#     'B': [100, 101, 99],
#     'C': [7, 8, 9]
# }
df = pd.DataFrame(data)
df = df.div(np.linalg.norm(df.values, axis=1),axis="index")
print(df)

mean = df.mean(axis="index")
sigma = np.sqrt(df.var(axis="index"))

norms = df.sub(mean,axis="columns")
norms = norms.div(sigma, axis="columns")
norms.sum(axis="columns")
```

```python
# L1-normalize

import pandas as pd
import numpy as np

# 
data = {
    'A': [100, 2, 3],
    'B': [10000, 101, 99],
    'C': [700, 8, 9]
}
# data = {
#     'A': [1, 2, 3],
#     'B': [100, 101, 99],
#     'C': [7, 8, 9]
# }
df = pd.DataFrame(data)
df = df.div(df.sum(axis="columns"),axis="index")
print(df)

mean = df.mean(axis="index")
sigma = np.sqrt(df.var(axis="index"))

norms = df.sub(mean,axis="columns")
norms = norms.div(sigma, axis="columns")
norms.sum(axis="columns")
```

## Calculate fit for a sample Paper 

```python
norms = pd.read_pickle(result_path+"norms_l1.pickle")
mean= pd.read_pickle(result_path+"mean.pickle")
sigma= pd.read_pickle(result_path+"simga.pickle")
journal_source_target= pd.read_pickle(result_path+"journal_source_target.pickle")
cs= pd.read_pickle(result_path+"cs.pickle")

doi_df = pd.read_pickle(data_path+"paper_detail/doi.pickle")
journalname_df = pd.read_pickle(data_path+"id_names/journal_info.pickle")

doi_df.head(), journalname_df.head()
```

```python
# sample_paper_doi = ""
sample_paper_id = cs.sample().index[0]

print(sample_paper_id)
# citation_cs[sample_paper_id]

target_journal_distribution = journal_source_target.loc[sample_paper_id]["target_journal"].value_counts(normalize=True)

mean.name = "mean"
sigma.name = "sigma"

target_journal_distribution = pd.merge(mean,target_journal_distribution,left_index=True,right_index=True, how="left" ).fillna(0)
df = pd.merge(sigma,target_journal_distribution,left_index=True,right_index=True, how="left")

z_score = (df["target_journal"] - df["mean"])/ df["sigma"]
norms_valid = norms.loc[:,~norms.isna().any(axis=0)] 
z_score_valid = z_score[~z_score.isna()]
z_score_valid = z_score_valid / np.linalg.norm(z_score_valid.values)

z_score_valid.shape, norms_valid.shape

plt.hist(np.dot(norms_valid,z_score_valid),bins=np.linspace(-0.5,10,105))
plt.yscale("log")

max_fit_id = np.argmax(np.dot(norms_valid,z_score_valid))
fit = np.max(np.dot(norms_valid,z_score_valid))
bestfit_journal_id = norms_valid.iloc[max_fit_id].name
original_journal_id = journal_source_target.loc[sample_paper_id]["source_journal"].values[0]

print("best fit journal id is",bestfit_journal_id, "with fit value:" ,fit)
print("original journal published is ", original_journal_id, "with fit value:", np.dot(norms_valid.loc[original_journal_id],z_score_valid))
print("The journal name is ", journalname_df.loc[original_journal_id]["name"])
print("and",journalname_df.loc[bestfit_journal_id]["name"],", respectively.")
```
